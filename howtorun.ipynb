# How to Run a planet simulastion using Hsimulator

## Import necessary modules

```python
import numpy as np
import Hsimulator as H

%matplotlib inline
```

## Create initial orbit

```python
# position
x0 = np.array([1.,0,0])
# velocity
v0 = np.array([0,0.75,0])
# orbit
o = [H.orbit(x0,v0)]
```

### Create snapshot

```python
# initial time
t0 = 0.
# snapshot
snp = H.snapshot(t0,o)
```

### Run the simulation

```python
H.run(snap=snp)
```
